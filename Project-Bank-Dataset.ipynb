```python
def IQR():
    
    q1=np.percentile(data["age"],25)
    q2=np.percentile(data["age"],50)
    q3=np.percentile(data["age"],75)
    iqr=(q3-q1)
    outlier1=q1-1.5*(iqr)
    outlier2=q3+1.5*(iqr)
    return(outlier1,outlier2,iqr)
```
